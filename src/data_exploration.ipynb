```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt         #  importing some basic libraries to work with the dataset and for visualization



def load_data(filepath):
    """
    Load the dataset into a pandas DataFrame.
    """
    df = pd.read_csv('/workspaces/ts-aiml-datascience-challenge-coderabhishek2060/data/bank_churn.csv')
    return df

def check_missing_values(df):
    """
    Check for missing values in the DataFrame.
    """
    missing = df.isnull()
    print("Missing Values:\n", missing)

def generate_summary_statistics(df):
    """
    Generate summary statistics for key variables.
    """
    summary = df.describe()
    print("Summary Statistics:\n", summary)

def visualize_distributions(df):
    """
    Visualize distributions of age, balance, credit_score, and estimated_salary.
    
    Hint: use sns.histplot() 
    """
    plt.figure(figsize=(14, 8))

    # Histogram for 'age'
    plt.subplot(2, 2, 1)
    sns.histplot(df['age'], kde=True, color='skyblue', bins=10)
    plt.title('Age Distribution')
    plt.xlabel('Age')

    # Histogram for 'balance'
    plt.subplot(2, 2, 2)
    sns.histplot(df['balance'], kde=True, color='green', bins=10)
    plt.title('Balance Distribution')
    plt.xlabel('Balance')

    # Histogram for 'credit_score'
    plt.subplot(2, 2, 3)
    sns.histplot(df['credit_score'], kde=True, color='purple', bins=10)
    plt.title('Credit Score Distribution')
    plt.xlabel('Credit Score')

    # Histogram for 'estimated_salary'
    plt.subplot(2, 2, 4)
    sns.histplot(df['estimated_salary'], kde=True, color='red', bins=10)
    plt.title('Estimated Salary Distribution')
    plt.xlabel('Estimated Salary')

    # Adjust layout
    plt.tight_layout()
    plt.show()

def main():
    # Load data
    df = load_data("copythefilepath")

    # Check for missing values
    check_missing_values(df)
    
    # Generate summary statistics
    generate_summary_statistics(df)
    
    # Visualize distributions
    visualize_distributions(df)

if __name__ == "__main__":
    main()
```
